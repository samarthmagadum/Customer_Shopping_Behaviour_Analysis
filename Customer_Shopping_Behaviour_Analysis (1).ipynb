```python
# Loading the dataset using pandas

import pandas as pd

df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Summary statistics using .describe()
df.describe(include='all')
```

```python
# Checking if missing data or null values are present in the dataset
df.isnull().sum()
```

```python
# Imputing missing values in Review Rating column with the median rating of the product category

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Renaming columns according to snake casing for better readability and documentation

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a new column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Dropping promo code used column

df = df.drop('promo_code_used', axis=1)
```

```python
from urllib.parse import quote_plus
import pandas as pd
from sqlalchemy import create_engine

# 1. SQL Server connection parameters for Windows Authentication
host = "localhost\SQLEXPRESS"  # Or your specific server name/SQLEXPRESS instance
database = "customer_behavior"

# 2. Build the connection string using Trusted_Connection=yes
# Note: your host string uses a semicolon (;) for parameters instead of the @ symbol
driver = quote_plus("ODBC Driver 17 for SQL Server")
connection_string = (
    f"mssql+pyodbc:///?odbc_connect="
    f"DRIVER={{{driver}}};"
    f"SERVER={host};"
    f"DATABASE={database};"
    f"Trusted_Connection=yes;"
)

engine = create_engine(connection_string)

# 3. Write DataFrame to SQL Server
df.to_sql("customer", engine, if_exists="replace", index=False)

# 4. Read back sample
result_df = pd.read_sql("SELECT TOP 5 * FROM customer;", engine)
print(result_df)
```
